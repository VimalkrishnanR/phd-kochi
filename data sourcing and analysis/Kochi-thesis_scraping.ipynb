# Scraping workflow notebook

```python
import requests
from bs4 import BeautifulSoup
import time
import pandas as pd
import io
```

# Getting list of Theses titles containing 'Kochi'

```python
url = "https://shodhganga.inflibnet.ac.in/simple-search?location=&query=Kochi&filter_field_1=title&filter_type_1=contains&filter_value_1=Kochi&filter_field_2=language&filter_type_2=equals&filter_value_2=English&rpp=60&sort_by=score&order=DESC&etal=0&submit_search=Update"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36" 
}
print ("time delay for 20 seconds...")
time.sleep(20)

print("Getting 60 titles with Kochi. This will take a while")
response = requests.get(url, headers=headers, timeout = 120)
if response.status_code == 200: 
    data = io.StringIO(response.text)
    table = pd.read_html(data)
#doc = BeautifulSoup(response.text, 'html.parser')
    print("success!")
else: 
    print(f"Status Code: {response.status_code}")
```

```python
type(table)
```

```python
kochi_title = pd.DataFrame(table[0])
```

```python
#kochi_title.head()
```

```python
kochi_title.to_csv("kochi_thesis_titles.csv", index = False)
```
